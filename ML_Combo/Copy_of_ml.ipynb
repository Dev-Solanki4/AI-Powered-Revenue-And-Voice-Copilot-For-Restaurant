```python
top_products = data['product_name'].value_counts().head(100).index
data = data[data['product_name'].isin(top_products)]
```

```python
basket = (data
          .groupby(['order_id','product_name'])['product_name']
          .count()
          .unstack()
          .fillna(0)
          .astype('int8'))
```

```python
import pandas as pd
# The error occurred because .apply() passes a Series to the lambda, making the boolean check ambiguous.
# We use a vectorized approach instead:
basket = (basket > 0).astype(int)
```

```python
basket = pd.pivot_table(
    data,
    index='order_id',
    columns='product_name',
    aggfunc='size',
    fill_value=0
)
```

```python
basket = basket.astype(bool)
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets = fpgrowth(
    basket,
    min_support=0.01,
    use_colnames=True
)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)

rules.head()
```

```python
rules = rules[
    (rules['confidence'] > 0.4) &
    (rules['lift'] > 1.2) &
    (rules['support'] > 0.01)
]
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
print(frequent_itemsets.head())
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules.sort_values(by="confidence", ascending=False)
rules.head(20)
```

```python
rules = rules.sort_values(by="confidence", ascending=False)
rules.head(20)
```

```python
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))
```

```python
rules['combo'] = rules['antecedents'] + " + " + rules['consequents']
rules[['combo','confidence','lift']].head(20)
```

```python
# Cluster products based on co-occurrence in orders and visualize clusters using Plotly.
# Plotly avoids matplotlib DLL issues in this environment and gives interactive output.
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd

try:
    import plotly.express as px
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "plotly"])
    import plotly.express as px

# Transpose the basket so rows are products and columns are orders (binary matrix)
product_matrix = basket.T.astype(float)

# Reduce dimensions for visualization
pca = PCA(n_components=2, random_state=42)
product_2d = pca.fit_transform(product_matrix)

# Choose number of clusters (tune as needed)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
product_labels = kmeans.fit_predict(product_2d)

# Build a DataFrame for plotting
plot_df = pd.DataFrame({
    'product': product_matrix.index,
    'x': product_2d[:, 0],
    'y': product_2d[:, 1],
    'cluster': product_labels,
})

fig = px.scatter(
    plot_df,
    x='x',
    y='y',
    color='cluster',
    hover_data=['product'],
    title=f'Product clusters (KMeans k={n_clusters})',
    labels={'x': 'PCA 1', 'y': 'PCA 2'},
)
fig.update_traces(marker=dict(size=10, opacity=0.75))
fig.show()

# Print example products per cluster
for c in sorted(plot_df['cluster'].unique()):
    sample = plot_df[plot_df['cluster'] == c]['product'].head(10).tolist()
    print(f"\nCluster {c} sample products ({len(sample)} shown):")
    print(sample)
```

```python
# Optionally visualize cluster centers (in PCA space)
centers = kmeans.cluster_centers_
centers_df = pd.DataFrame(centers, columns=['x', 'y'])
centers_df['cluster'] = range(len(centers_df))

fig = px.scatter(
    plot_df,
    x='x',
    y='y',
    color='cluster',
    hover_data=['product'],
    title=f'Product clusters + centers (k={n_clusters})',
    labels={'x': 'PCA 1', 'y': 'PCA 2'},
)
fig.add_scatter(
    x=centers_df['x'],
    y=centers_df['y'],
    mode='markers+text',
    marker=dict(size=18, symbol='x', color='black'),
    text=centers_df['cluster'],
    textposition='top center',
    name='centers',
)
fig.show()
```

```python
# Save trained clustering model + PCA transformer + product index so a web app can load and use it.
import pickle

model_artifact = {
    "kmeans": kmeans,
    "pca": pca,
    # Needed to map web-input product names to the expected feature order
    "product_index": list(product_matrix.index),
    "n_clusters": n_clusters,
}

with open("product_cluster_model.pkl", "wb") as f:
    pickle.dump(model_artifact, f)

print("Saved product_cluster_model.pkl (contains kmeans+pca+product_index)")
```
